```python
import nltk
def unusual_words(text):
    text_vocab = set(w.lower() for w in text if w.isalpha())
    english_vocab = set(w.lower() for w in nltk.corpus.words.words())
    #unusual = text_vocab - english_vocab
    unusual = text_vocab.difference(english_vocab)
    
    return sorted(unusual)
```

```python
unusual_words(nltk.corpus.gutenberg.words('austen-sense.txt'))
```

```python
unusual_words(nltk.corpus.nps_chat.words())
```

```python
from nltk.corpus import brown
def content_fraction(text):
    stopwords = nltk.corpus.stopwords.words('english')
    content = [w for w in text if w.lower() not in stopwords]
    return len(content) / len(text)
```

```python
content_fraction(nltk.corpus.brown.words(categories = 'news'))
```

```python
content_fraction(nltk.corpus.reuters.words())
```

```python
def stress(pron):
    result = []
    
    for phone in pron:
        for char in phone:
            if char.isdigit():
                result.append(char)
    return result

rule = ['0','1','0','2']
for w, pron in entries:
    if stress(pron) == rule:
        print(w, pron)
```

```python
T4 = [(pron[0]+'--'+pron[3],word)
      for (word,pron) in entries
      if pron[0] == 'T' and len(pron) == 4]

cfd = nltk.ConditionalFreqDist(T4)
for template in sorted(cfd.conditions()):
    if len(cfd[template]) > 10:
        cfd_temp = nltk.ConditionalFreqDist(cfd[template].most_common(10))
        words = sorted(cfd_temp)
        # print(words)
        wordlist = ' '.join(words)
        print(template,wordlist)
```
